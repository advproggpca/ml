```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X= np.arange(1, 11, 0.5)
y = 3*X - 1
X, y
```

```python
X_train = X[:15]
X_test = X[15:]
y_train = y[:15]
y_test = y[15:]
```

```python
X_train, y_train, X_test, y_test
```

```python
def plot_data(X_train = X_train, y_train = y_train, X_test = X_test, y_test = y_test, y_pred = None):    
    plt.figure(figsize=(10, 4))
    plt.scatter(X_train, y_train, c="red",  label="Train Data") 
    plt.scatter(X_test, y_test, c="blue", label="Test Data")
    if y_pred is not None:
        plt.scatter(X_test , y_pred, c="green", label="Predicted Data")
    plt.legend(prop={"size": 14});
```

```python
plot_data(X_train, y_train, X_test, y_test)
```

```python
weight = 0.7
bias = 0.3
```

```python
y_test_pred = weight * X_test + bias
```

```python
y_train_pred = weight * X_train + bias
```

```python
X_train,y_train_pred
```

```python
plot_data(X_train, y_train, X_test, y_test, y_train_pred)
```

```python
wl = []
bl = []
Ll = []
lr = 0.01
for _ in range(1000):
    y_pred = weight * X_train + bias

    error = y_pred - y_train
    dw = np.mean(2 * error * X_train)
    db = np.mean(2 * error)

    weight -= lr * dw
    bias -= lr * db

    L = np.mean((y_train - y_pred)**2)

    wl.append(weight)
    bl.append(bias)
    Ll.append(L)

    if _ % 10 == 0:
        print(f"iter {_}: weight={weight}, bias={bias}, loss={L}")
```

```python
weight, bias, L
```

```python
# Loss curve
plt.figure()
plt.plot(Ll, range(0,1000), label="Loss")
plt.title("Loss vs Iterations")
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.legend()
plt.show()

# Weight convergence
plt.figure()
plt.plot(wl)
plt.title("Weight convergence")
plt.xlabel("Iterations")
plt.ylabel("Weight")
plt.show()

# Bias convergence
plt.figure()
plt.plot(bl)
plt.title("Bias convergence")
plt.xlabel("Iterations")
plt.ylabel("Bias")
plt.show()
```

```python
plot_data(X_train, y_train, X_test, y_test, y_train_pred)
```

```python
y_test_pred = weight * X_test + bias
y_test_pred, y_test
```

```python
plot_data(X_train, y_train, X_test, y_test, y_test_pred)
```

```python
test_loss = np.mean((y_test - y_test_pred)**2)
print(f"Test Loss: {test_loss}")
```

```python
train_loss = np.mean((y_train - y_train_pred)**2)
print(f"Train Loss: {train_loss}")
```
