```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e:
        print(e)
```

```python
# Create a simple model
model = keras.Sequential([
    layers.Dense(16, activation='relu', input_shape=(10,)),  # input layer
    layers.Dense(8, activation='relu'),                      # hidden layer
    layers.Dense(1, activation='sigmoid')                    # output layer
])
```

```python
# Compile
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
# Dummy data
import numpy as np
X = np.random.random((100, 10))
y = np.random.randint(0, 2, (100, 1))
```

```python
X,y
```

```python
# Train
model.fit(X, y, epochs=5, batch_size=8)
```

```python
predictions = model.predict(X)
predicted_classes = (predictions > 0.5).astype(int)

print("Predicted probabilities:", predictions[:10].flatten())
print("Predicted classes:", predicted_classes[:10].flatten())
```
